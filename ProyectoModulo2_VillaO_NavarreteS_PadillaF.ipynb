Las ecuaciones Lotka-Volterra, también conocindas como ecuaciones presa-depredador, son un par de ecuaciones diferenciales de primer orden que se utilizan normalmente para describir dinámicas naturales en las cuales se enfrenten dos especies, una tomando el rol de presa, mientras la contraparte juega como depredador.

# 1.3 Modelo

En un escenario cualquiera, dígase la Sabana Africana, se encuentran dos especies, los antílopes y los chitas. Para motivos prácticos, partimos de los siguientes supuestos:

- Son las únicas 2 especies viviendo en esa determinada zona.
- No existen enfermedades o plagas que pudieran afectar a las especies.
- La interdependencia se definirá cada caso.

Sean:
- $x_1$: Antílopes.
- $x_2$: Chitas.

El modelo descrito es el siguiente
\begin{align}
\frac{dx_1}{dt}&=x_1(a-b\;x_2)\\
\frac{dx_2}{dt}&=-x_2(c-d\;x_1),
\end{align}
Definimos $x=\left[x_1\quad x_2\right]^T$.
- $a$ = Tasa de crecimiento de las presas
- $b$ = Éxito en la caza del depredador
- $c$ = Tasa de crecimiento de los depredadores
- $d$ = Tasa de éxito para alimentarse del depredador

```python
# Primer Caso: a,b,c,d positivas e iguales

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def nat1(x,t,a,b,c,d):
    x1= x[0]
    x2= x[1]
    return [x1*(a-b*x2), -x2*(c-d*x1)]

x0= [1.5,1]

t = np.linspace(0,15,100)

a,b,c,d = 1,1,1,1
```

```python
x= odeint(nat1, x0, t, args=(a,b,c,d))
x1 = x[:,0]
x2 = x[:,1]
```

```python
plt.figure(figsize=(8,6))
plt.plot(t, x1, 'r', lw=2, label='$x_1(t)$: comportamiento de los antílopes')
plt.plot(t, x2, 'b', lw=2, label='$x_2(t)$: comportamiento de los chitas')
plt.grid()
plt.legend(loc='best')
plt.xlabel('T [unidades de tiempo]')
plt.ylabel('Población [miles]')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(x1, x2, lw=2)
plt.grid()
plt.xlabel('$x_1$ comportamiento de los antílopes')
plt.ylabel('$x_2$ comportamiento de los chitas')
plt.show()
```

Ahora bien, ¿Qué sucede cuando la tasa de crecimiento de la presa es significativamente mayor a la de los depredadores?

```python
def nat2(x,t,a,b,c,d):
    x1= x[0]
    x2= x[1]
    return [x1*(a-b*x2), -x2*(c-d*x1)]

x0= [1.5,1]

t = np.linspace(0,15,200)

a,b,c,d = 3,1.1,1.2,.8
```

```python
x= odeint(nat2, x0, t, args=(a,b,c,d))
x1 = x[:,0]
x2 = x[:,1]
```

```python
plt.figure(figsize=(8,6))
plt.plot(t, x1, 'r', lw=2, label='$x_1(t)$: comportamiento de los antílopes')
plt.plot(t, x2, 'b', lw=2, label='$x_2(t)$: comportamiento de los chitas')
plt.grid()
plt.legend(loc='best')
plt.xlabel('T [unidades de tiempo]')
plt.ylabel('Población [miles]')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(x1, x2, lw=2)
plt.grid()
plt.xlabel('$x_1$ comportamiento de los antílopes')
plt.ylabel('$x_2$ comportamiento de los chitas')
plt.show()
```

¿Qué sucederá si la tasa de crecimiento de la especie que representa a la presa es menor y el índice de éxito de los depredadores al cazar es alto?

```python
def nat3(x,t,a,b,c,d):
    x1= x[0]
    x2= x[1]
    return [x1*(a-b*x2), -x2*(c-d*x1)]

x0= [1.5,1]

t = np.linspace(0,15,200)

a,b,c,d = .8,1,.2,2
```

```python
x= odeint(nat3, x0, t, args=(a,b,c,d))
x1 = x[:,0]
x2 = x[:,1]
```

```python
plt.figure(figsize=(8,6))
plt.plot(t, x1, 'r', lw=2, label='$x_1(t)$: comportamiento de los antílopes')
plt.plot(t, x2, 'b', lw=2, label='$x_2(t)$: comportamiento de los chitas')
plt.grid()
plt.legend(loc='best')
plt.xlabel('T [unidades de tiempo]')
plt.ylabel('Población [miles]')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.plot(x1, x2, lw=2)
plt.grid()
plt.xlabel('$x_1$ comportamiento de los antílopes')
plt.ylabel('$x_2$ comportamiento de los chitas')
plt.show()
```
